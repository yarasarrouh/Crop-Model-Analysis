```python
import arc
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import os
import json

# Based on Yin Feng code only the loop was added and the constants were chosen to be adequate for potatoes. 

# Constants
START_OF_SEASON = 138
CROP_TYPE = 'maize'
NUM_SAMPLES = 100000
GROWTH_SEASON_LENGTH = 90
LAZY_EVALUATION_STEP = 100
ALPHA = 0.8
LINE_WIDTH = 2
start_date="2022-05-18"
end_date="2022-11-02"

# Variables
folder_path = Path.home() / 'Downloads/HI_Data_yara_new/code_data/save_code_data/fields_boundaries_json_2022'
output_folder = Path.home() / 'Downloads/2022-UCL-model_output_maize'

# Loop over all files in the folder
for filename in os.listdir(folder_path):
    geojson_path = os.path.join(folder_path, filename)
    
    # Generate the output folder path for this file
    output_file_path = output_folder / f'{Path(geojson_path).stem}/SF_field.npz'
    
    # Check if the output file already exists, if yes, skip this iteration
    if output_file_path.exists():
        print(f"Output file already exists for {filename}. Skipping.")
        continue
    
    # Print the filename before processing
    print(f"Processing file: {filename}")
    
    # Downloading the S2 corresponding images
    arc_dir = os.path.dirname(os.path.realpath(arc.__file__))
    S2_data_folder = output_folder / Path(geojson_path).stem
    S2_data_folder.mkdir(parents=True, exist_ok=True)
    
    scale_data, post_bio_tensor, post_bio_unc_tensor, mask, doys = arc.arc_field(
        start_date, 
        end_date, 
        geojson_path, 
        START_OF_SEASON, 
        CROP_TYPE, 
        str(output_file_path), 
        NUM_SAMPLES, 
        GROWTH_SEASON_LENGTH, 
        str(S2_data_folder),
        plot=True
    )
```
